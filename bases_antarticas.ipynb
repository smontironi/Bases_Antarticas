```python
#esto funciona en colab
import pandas as pd
url = 'https://www.ign.gob.ar/NuestrasActividades/Geografia/DatosArgentina/BasesAntartida'
bases_antarticas = pd.read_html(url)
```

```python
bases_temporarias= bases_antarticas[0]
```

```python
#agrego la columna tipo para aclarar si es temporal o permanente
bases_temporarias_final = bases_temporarias.assign(Tipo="Temporaria")
```

```python
bases_permanentes= bases_antarticas[1]
```

```python
#del mismo modo que las temporales hago lo mismo con las permanentes
#agrego la columna tipo para aclarar si es temporal o permanente
bases_permanentes_final = bases_permanentes.assign(Tipo="Permanentes")
```

```python
#uno los dos dataframes para generar un unico df
bases_final = pd.concat([bases_temporarias_final, bases_permanentes_final], ignore_index=True)
```

```python
bases_final
```

```python
import re

# Función para convertir DMS a grados decimales
def dms_a_decimal(dms):
    # Expresión regular para extraer grados, minutos y segundos
    dms_regex = re.compile(r"(\d+)[°º](\d+)'(\d+)\"([NSEWO])")
    match = dms_regex.match(dms)

    if match:
        grados, minutos, segundos, direccion = match.groups()
        decimal = float(grados) + float(minutos)/60 + float(segundos)/3600
        if direccion in ['S', 'O']:
            decimal = -decimal
        return decimal
    else:
        raise ValueError("Formato DMS no válido")

# Función para separar las coordenadas en latitud y longitud
def extraer_latitud_longitud(coord):
    lat_dms, lon_dms = coord.split(' ')
    lat = dms_a_decimal(lat_dms)
    lon = dms_a_decimal(lon_dms)
    return pd.Series([lat, lon])

# Aplicar la función al DataFrame
bases_final[['Latitud', 'Longitud']] = bases_final['Coordenadas'].apply(extraer_latitud_longitud)
```
